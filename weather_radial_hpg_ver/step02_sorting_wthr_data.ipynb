```python
import json
from datetime import datetime
import time
import pandas as pd
```

```python
# Source : http://stackoverflow.com/questions/26595647/store-and-recall-a-variable-in-a-text-file-python
source = open('data.txt', 'r').read()
data = json.loads(source)
```

```python
len(data)
```

```python
data[1]
```

```python
weatherColumbia = []
for i in range(0,365):
    bucket = {}
    daily_data = data[i]['daily']['data'][0]
    daily_data_date = datetime.fromtimestamp(int(daily_data['time'])).strftime('%Y-%m-%d')
    bucket['date']=daily_data_date
    bucket['max_temp']= daily_data['temperatureMax']
    bucket['mean_temp'] = (daily_data['temperatureMax']+daily_data['temperatureMin'])/2
    bucket['min_temp']= daily_data['temperatureMin']
    weatherColumbia.append(bucket)
    
```

```python
weatherColumbia
```

```python
df = pd.DataFrame(weatherColumbia)
df.head()
```
