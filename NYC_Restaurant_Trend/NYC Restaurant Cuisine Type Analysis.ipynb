```python
import pandas as pd                  
import datetime as dt    
import matplotlib.pyplot as plt                       
import numpy as np  
%matplotlib inline             
```

```python
# 2016
# Due to continuous crash, the data limited to 50000
url_2016= "https://data.cityofnewyork.us/resource/xx67-kt59.json?&$limit=50000&$select=boro,cuisine_description"
df_2016= pd.read_json(url_2016)
type_2016 = df_2016.groupby('cuisine_description').size().sort_values()
```

```python
df_2016.describe()
```

```python
items_counts = df_2016['cuisine_description'].value_counts()
items_counts.head()
```

```python
fig, ax = plt.subplots()
items_counts.head(10).plot(ax=ax,kind="barh",figsize=(5, 5))
ax.set_title('Restaurants - Cuisine Type 2016')
ax.set_xlabel('# of Restaurants')
ax.set_ylabel('Cuisine type')
```

```python
# 2015 
url_2015="https://data.cityofnewyork.us/resource/vv3s-su4i.json?&$limit=50000&$select=boro,cuisine_description"
df_2015= pd.read_json(url_2015)
type_2015 = df_2015.groupby('cuisine_description').size().sort_values()
items_counts_2015 = df_2015['cuisine_description'].value_counts()
items_counts_2015.head(10)
```

```python
fig, ax = plt.subplots()
items_counts_2015.head(10).plot(ax=ax,kind="barh",figsize=(5, 5))
ax.set_title('Restaurants - Cuisine Type 2015')
ax.set_xlabel('# of Restaurants')
ax.set_ylabel('Cuisine type')
```

```python
# boro analysis

url_bor = "https://data.cityofnewyork.us/resource/batu-qkuq.json?&$limit=50000&$select=camis,boro,cuisine_description"
df_bor = pd.read_json(url_bor)
grouped_bor = df_bor.groupby(["boro", "cuisine_description"]).count()
fig, ax = plt.subplots()
grouped_bor.plot(ax=ax,kind="barh",figsize=(22,50))
ax.set_title('Cuisine Types', loc='left')
ax.set_xlabel("# of Restaurants")
ax.set_ylabel('Cuisine Type')
ax.legend_.remove()
```
